```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Baseline and All ML algorithms

```python
results = pd.DataFrame([
    {"model": "Majority Baseline", "accuracy": 0.201, "macro-f1": 0.048},
    {"model": "Random Baseline", "accuracy": 0.171, "macro-f1": 0.167},

    {"model": "Linear SVM", "accuracy": 0.206, "macro-f1": 0.192},
    {"model": "Logistic Regression", "accuracy": 0.221, "macro-f1": 0.212},
    {"model": "FFNN", "accuracy": 0.246, "macro-f1": 0.179},

    {"model": "FFNN+skip", "accuracy":0.271, "macro-f1":  0.232},
    {"model": "Random Forest", "accuracy": 0.312, "macro-f1": 0.259}
])

x = np.arange(len(results))
width = 0.35

plt.figure()
plt.barh(x - width/2, results["accuracy"],  width, label="Accuracy")
plt.barh(x + width/2, results["macro-f1"], width, label="Macro-F1")

for i, v in enumerate(results["accuracy"]):
    plt.text(v + 0.001, i - width/2, f"{v}", va='center',fontsize=8)

for i, v in enumerate(results["macro-f1"]):
    plt.text(v + 0.001, i + width/2, f"{v}", va='center',fontsize=8)

plt.yticks(x, results["model"])
plt.xlabel("Score")
plt.xlim(0, max(results["accuracy"].max(), results["macro-f1"].max()) + 0.03)
plt.legend()
plt.gca().invert_yaxis()
plt.savefig("../outputs/figs/performance_all_models.png",bbox_inches='tight', dpi=300)
```

### FFNN+skip and RF optimised

```python
results_2 = pd.DataFrame([
    {"model": "FFNN+skip", "accuracy":0.271, "macro-f1":  0.232},
    {"model": "FFNN+skip (with participant_id)", "accuracy": 0.281, "macro-f1": 0.256},
    {"model": "Feature Selection + FFNN+skip", "accuracy": 0.266, "macro-f1": 0.246},


    {"model": "Random Forest", "accuracy": 0.312, "macro-f1": 0.259},
    {"model": "Random Forest (with participant_id)", "accuracy": 0.281, "macro-f1": 0.235},
    {"model": "Feature Selection + Random Forest", "accuracy": 0.327, "macro-f1": 0.286}
])

x = np.arange(len(results_2))
width = 0.35

plt.figure()
plt.barh(x - width/2, results_2["accuracy"],  width, label="Accuracy")
plt.barh(x + width/2, results_2["macro-f1"], width, label="Macro-F1")
plt.axhline(2.5, color="gray", linestyle="--", linewidth=1)

for i, v in enumerate(results_2["accuracy"]):
    plt.text(v + 0.001, i - width/2, f"{v}", va='center',fontsize=8)

for i, v in enumerate(results_2["macro-f1"]):
    plt.text(v + 0.001, i + width/2, f"{v}", va='center',fontsize=8)

plt.text(0.385, 2.4, "FFNN+skip", fontsize=8, fontweight="bold", color='gray')
plt.text(0.36, 2.75, "Random Forest", fontsize=8, fontweight="bold", color='gray')

plt.yticks(x, results_2["model"], fontsize=10)
plt.xlabel("Score")
plt.xlim(0, max(results_2["accuracy"].max(), results_2["macro-f1"].max()) + 0.12)
plt.legend()
plt.gca().invert_yaxis()

plt.savefig("../outputs/figs/id_and_selection.png",bbox_inches='tight', dpi=300)
```
